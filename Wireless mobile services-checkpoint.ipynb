Topic modelling: Mint Mobile

```python
#Tokenization #removing the stopwords and Punctations 
#The goal of MIN_DF is to ignore words that have very few occurrences to be considered meaningful. 
#For example, in your text you may have names of people that may appear in only 1 or two documents
#Similarly, you can ignore words that are too common with MAX_DF.
#Max_df on 1 because I already removed the stopwords! if it's found in all documents (100% or 1.0) then it's ignored."
import gensim
from gensim.models.wrappers import LdaMallet
import os
import gensim

import pyLDAvis
import pyLDAvis.gensim_models

from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(
analyzer='word',       
min_df=3,# minimum required occurences of a word 
stop_words='english',# remove stop words
lowercase=True,# convert all words to lowercase
token_pattern='[a-zA-Z0-9]{3,}',# num chars > 3
max_features=5000,# max number of unique words
                            )
data_matrix = vectorizer.fit_transform(mint.tweet)
data_matrix
```

```python
from sklearn.decomposition import LatentDirichletAllocation

lda_model = LatentDirichletAllocation(
n_components=8, # Number of topics
learning_method='online',
random_state=20,       
n_jobs = -1  # Use all available CPUs
                                     )

lda_output = lda_model.fit_transform(data_matrix)
```

```python
import pyLDAvis
import pyLDAvis.sklearn
pyLDAvis.enable_notebook()
pyLDAvis.sklearn.prepare(lda_model, data_matrix, vectorizer, mds='tsne')
```

```python
#0 User satisfaction
#1 Affordability
#2 Willingness
```

```python
for i,topic in enumerate(lda_model.components_):
    print(f'Top 10 words for topic #{i}:')
    print([vectorizer.get_feature_names()[i] for i in topic.argsort()[-10:]])
    print('\n')
```

```python
topic_values = lda_model.transform(data_matrix)
mint['Topic'] = topic_values.argmax(axis=1)
```

Sentiment analysis

```python
sentdf=pd.concat([usersat,afford,willing,usersat1,afford1,willing1])
```

```python
import seaborn as sns
from IPython.display import display
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.sentiment.util import *
from textblob import TextBlob
from nltk import tokenize
sentdf.head()
```

```python
def get_polarity(text):
    return TextBlob(text).sentiment.polarity
sentdf['Polarity'] = sentdf['tweet'].apply(get_polarity)
```

```python
sentdf['Sentiment_Type']=''
sentdf.loc[sentdf.Polarity>0,'Sentiment_Type']='POSITIVE'
sentdf.loc[sentdf.Polarity==0,'Sentiment_Type']='NEUTRAL'
sentdf.loc[sentdf.Polarity<0,'Sentiment_Type']='NEGATIVE'
sentdf.head()
```
